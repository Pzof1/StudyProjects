```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="mitchelldehaven/whisper-large-v2-ru")
```

```python
# Load model directly
from transformers import AutoProcessor, AutoModelForSpeechSeq2Seq

processor = AutoProcessor.from_pretrained("mitchelldehaven/whisper-large-v2-ru")
model = AutoModelForSpeechSeq2Seq.from_pretrained("mitchelldehaven/whisper-large-v2-ru")
```

```python
# Обработка аудиофайла
result = pipe('test1.mp3', chunk_length_s=30)  # chunk_length_s — длина фрагмента для обработки

# Вывод результата
print(result["text"])
```
